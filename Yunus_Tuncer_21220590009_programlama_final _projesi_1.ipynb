* Veri setinin son sütununda bulundurduğumuz y - x koordinatlarını okuyalım,

* Daha sonra Projeksiyon ayarlarını yapalım,

* Ve son olarak csv dosyasını geopandas kullanarak shp dosyasına çevirelim.

```python
geometry = [Point(xy) for xy in zip(df.x, df.y)]
crs = {'init': 'epsg:4326'} 
geo_df = GeoDataFrame(df, crs=crs, geometry=geometry)

geo_df.to_file(driver='ESRI Shapefile', filename='Batman_Merkez_Petrol_İstasyonları.shp')
```
