* yp_sırala adında yeni bir değişken oluşturalım.

* yp_uygulamalar veri setini değerlendirmeye göre filtreleyelim ve NumPy dizisi haline getirip değişken olarak atıyalım.

* Bu dizinin max ve min değerlerini inceliyelim

* bu içerikleri oy değerlerini minimum değerden bir eksili ve maxsimum değerden üç fazlasını alıp bir bir artacak şekilde dağılımı sağlayan bir matplotlib histogramı çizdirelim.

.

* **özelleştirme**

* histogram boyutunu 9,6 olarak ayarlıyalım

* Grafikteki barları kırmızı, renk şefaflığını 0.72 ve bar çevresini siyah olarak ayarlıyalım.

* histogram başlığını "PLAY STORDEKİ OYLAR VERİLME ORANLARI", yazı boyunu 12 ve  kalın olarak ayarlayalım.


* Grafiğin x sütununa "Adeti" ,y sütununa "oy aralığı" olarak yazıp, yazı  boyunu 12 olarak ayarlıyalım

```python
yp_sırala = yp_uygulamalar['Rating'].to_numpy()

print(type(yp_sırala))
print(min(yp_sırala))
print(max(yp_sırala))

num_bins= np.arange(min(yp_sırala)-1, max(yp_sırala) + 3, 1)


plt.figure(figsize = (9,6))

plt.hist(yp_sırala, num_bins, facecolor='red', alpha=0.72, edgecolor='black') 

plt.title("\nPLAY STORDEKİ OYLAR VERİLME ORANLARI\n"  , weight='bold', size=12 ) 

plt.ylabel('Adeti\n' , size=12)
plt.xlabel('oy Aralığı' , size=12 )



plt.show()
```
